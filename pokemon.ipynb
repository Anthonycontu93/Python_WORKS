```python
import pandas as pd 
path_dataset2 = "C:/Users/anthony/Desktop/db_esercitazione/dataset/beginner_datasets/"
file2 = "pokemon.csv"
path2 = path_dataset2 + file2

poke = pd.read_csv(path2)

poke.sample(10)


# aggiungi indice che era colonna 
poke.nunique()

poke.columns



#poke.set_index("#", inplace = True)

# filtro leggenda ed grass 
filtro_legend = poke.Legendary
# uso ["Type 1 "] perchÃ¨ c'Ã¨ uno spazio 
filtro_grass = poke["Type 1"] == "Grass"

poke.loc[filtro_legend & filtro_grass]

filtro_ice = poke["Type 1"] == "Ice"
filtro_fire = poke["Type 1"] == "Fire"

filtro_complex = filtro_legend & (filtro_ice | filtro_fire)
poke.loc[filtro_complex]


poke.set_index("Name", inplace = True)

filtro_gen = poke.Generation == 1
filtro_attack = poke.Attack > 50
filtro_hp = poke.HP < 60 

filtro_com = filtro_gen & filtro_attack & filtro_hp 

poke.loc[filtro_com]




```
