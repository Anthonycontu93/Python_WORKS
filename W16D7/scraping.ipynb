```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd 

url = "https://finance.yahoo.com/crypto"

# prendo url della pagina 
page = requests.get(url) 
page
```

```python
# utilizzo la lib beautiful soup per stampare a video la pagina html in testo
soup = BeautifulSoup(page.text)
tokens = []
# con oggetto soup e metodo find all mi prendo solo ciò che cerco, la classe che sta tra i <td> (in questo caso)
# utilizzo {"class": e attributi} potrei utilizzare anceh class_ ma cos' solo con codice html standard 
lista_token = soup.find_all("td", attrs = {"class":"Va(m) Ta(start) Px(10px) Fz(s)"}) #aria-label="Name">Bitcoin USD</td>
```

```python
# mi prendo i toekns dalla lista oggetto di beautifulsoup 
for token in lista_token:
    # token.text mi prendo solo le var che mi servono 
    tokens.append(token.text)

tokens
```

```python
# non funziona per il market cap posso provare a prendere la classe padre poichè questa è dinamica 
lista_mcap = soup.find_all("span", attrs ={"class":"e3b14781 dde7f18a"})
lista_mcap
```

```python
#market cap preso ma con molti più oggetti 
caps = []
lista_mcap = soup.find_all("td", attrs ={"class":"Va(m) Ta(end) Pstart(20px) Fz(s)"})
for cap in lista_mcap:
    caps.append(cap.text)

caps
```

```python
#slice elementi mi servono solo ogni 5 
caps = caps[::5]
caps
```

```python
prices = []
lista_price = soup.find_all("td", attrs ={"class":"Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)"})
for p in lista_price:
    prices.append(p.text)

prices
```

```python
# me ne prendo 1 ogni 3 e pulisco gli elementi dalla virgola o dal +
prices_ = []
for price in prices[::3]:
    prices_.append(price.replace(",", ""))
    
changes_24 = []
for volatilita in prices[1::3]:
    changes_24.append(volatilita.replace("+", ""))
    
per_cambio = prices[2::3]
```

```python
# creo dataframe con un dict delle liste appena create dal web 

crypto_dict = {

    "token": tokens,
    "market_cap": caps,
    "price": prices_,
    "cambio24h":changes_24,
    "% 24h": per_cambio
}

crypto = pd.DataFrame(crypto_dict)

crypto
```
