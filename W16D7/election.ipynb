```python
# prendo tutte le righe ma solo le colonne indicate
candidati = elect.loc[:, ["Coderre", "Bergeron", "Joly"]]
candidati
```

```python
#somma delle righe per ogni colonna 
candidati.sum()
```

```python
sum_voti = candidati.sum()

sum_voti.plot(kind="bar")
```

```python
# prendo le colonne distrct e total e tutte le righe 
district_sum = elect.loc[:,["district", "total"]]
# assegno indice 
district_sum = district_sum.set_index("district")
```

```python
# totale voti per distretto
#figsize allarga le righe e colonne del grafico
district_sum.plot(kind = "barh", figsize=[10,15], grid = True)
```
