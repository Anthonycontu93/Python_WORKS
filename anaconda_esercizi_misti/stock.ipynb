```python
import pandas as pd 

stock = pd.read_csv("./beginner_datasets/stockdata.csv")
```

```python
# qui vedo già il df trasformato in realtà il dataframe non ha indice date e nemmeno colonna year
stock.head()
```

```python
# trasformo il df con la colonna Date in indice 
stock = stock.set_index("Date")
```

```python
stock.AAPL.plot()
```

```python
# plot per tutte le colonne tranne le ultime 2 
stock.iloc[:, :-2].plot()
```

```python
stock.columns
```

```python
# trasformo l'index in colonna date into a colonna year per anno
stock["year"] = pd.to_datetime(stock.index)
stock.year = stock.year.dt.year

# creo filtro per anno 2012 2013 
filtro = (stock.year == 2013) | (stock.year == 2012)
# con loc cerco sul df le righe che rispettano il filtro e le colonne che passo nella lista 
stock.loc[filtro, ["MSFT", "AAPL"]].plot()
```

```python
# grafico per plottizzare il valore delle azioni Microsoft 
stock.MSFT.plot(kind="hist")
```
